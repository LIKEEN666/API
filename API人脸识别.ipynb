# 1、使用Azure平台

* 情绪判别。

* 通过人脸的笑容弧度等各项值综合起来判别，判别这个人的情绪究竟是开心、伤心或者生气之类的，帮助那些无法判别别人情绪的人。通过使用该功能去获得他/她想要了解的那个人的情绪。(例如像《生活大爆炸》里谢尔顿那样无法判别别人情绪的人。)

```python
import requests
import json

# set to your own subscription key value
subscription_key = '0d71bf2686c2439eb12fa197e429d738'
assert subscription_key

# replace <My Endpoint String> with the string from your endpoint URL
face_api_url = 'https://westcentralus.api.cognitive.microsoft.com/face/v1.0/detect'

image_url = 'https://timgsa.baidu.com/timg?image&quality=80&size=b9999_10000&sec=1585765136436&di=aef89ecc5be743334c935c39ce781aff&imgtype=0&src=http%3A%2F%2Fci.xiaohongshu.com%2F74db99d8-c7a9-468f-8c62-e4f1269c853f%40r_640w_640h.jpg'
headers = {'Ocp-Apim-Subscription-Key': subscription_key}

params = {
    'returnFaceId': 'true',
    'returnFaceLandmarks': 'false',
    'returnFaceAttributes': 'age,gender,smile,glasses,emotion',
}

r = requests.post(face_api_url, params=params,
                         headers=headers, json={"url": image_url})
print(json.dumps(r.json()))
results = r.json()
results
```

```python
import pandas as pd
df = pd.io.json.json_normalize(results)
df
```

```python

df = df.rename ( columns = {"faceAttributes.age":"年龄",
                                 "faceAttributes.emotion.anger":"生气",
                                 "faceAttributes.emotion.contempt":"轻蔑",
                                 "faceAttributes.emotion.disgust":"厌恶",
                                 "faceAttributes.emotion.fear":"害怕",
                                 "faceAttributes.emotion.happiness":"高兴",
                                 "faceAttributes.emotion.neutral":"平静",
                                 "faceAttributes.emotion.sadness":"伤心",
                                 "faceAttributes.emotion.surprise":"惊讶",
                                 "faceAttributes.gender":"性别",
                                 "faceAttributes.glasses":"眼镜",
                                 "faceAttributes.smile":"微笑",
                                 })
df
       
```
