```python
# 封装成函数方便添加多张图片
def AddFace(img_url=str,userData=str):
    
    add_face_url ="https://api666.cognitiveservices.azure.com/face/v1.0/facelists/{faceListId}/persistedFaces"
    subscription_key = "2970fc774b7b4c9aafe8e4e610b57d09"
    assert subscription_key
    headers = {
        # Request headers
        'Content-Type': 'application/json',
        'Ocp-Apim-Subscription-Key': subscription_key,
    }
    img_url = "https://timgsa.baidu.com/timg?image&quality=80&size=b9999_10000&sec=1586756677795&di=7c08896af04221f764774f72abcbce33&imgtype=0&src=http%3A%2F%2F5b0988e595225.cdn.sohucs.com%2Fimages%2F20180122%2Fb15581cc9fa548baa6083c68034a2690.jpeg"

    params_add_face={
        "faceListId":"li_01",
        "userData": userData
     }

    r_add_face = requests.post(add_face_url,headers=headers,params=params_add_face,json={"url":img_url})
    return r_add_face.status_code
```

## 检测人脸

```python
# replace <My Endpoint String> with the string from your endpoint URL
face_api_url = 'https://api666.cognitiveservices.azure.com/face/v1.0/detect'

# 请求正文
image_url = 'https://ss1.bdstatic.com/70cFvXSh_Q1YnxGkpoWK1HF6hhy/it/u=3756097749,2266463588&fm=26&gp=0.jpg'

headers = {'Ocp-Apim-Subscription-Key': subscription_key}

# 请求参数
params = {
    'returnFaceId': 'true',
    'returnFaceLandmarks': 'false',
    # 选择model
    'recognitionModel':'recognition_01',
    'detectionModel':'detection_01',
    # 可选参数,请仔细阅读API文档
    'returnFaceAttributes': 'age,gender,headPose,smile,facialHair,glasses,emotion,hair,makeup,occlusion,accessories,blur,exposure,noise',
}

response = requests.post(face_api_url, params=params,
                         headers=headers, json={"url": image_url})
# json.dumps 将json--->字符串
json.dumps(response.json())
```

```python
胡歌 = response.json()[0]["faceId"]
```
